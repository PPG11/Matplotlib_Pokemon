### 常用

```python
x = np.linspace(0.05, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='sin(x)')
# xy轴标签
plt.xlabel('variable x');
plt.ylabel('value y');
# 标题
plt.title('sin x');
# 水平参考线
plt.axhline(y=0.8, ls='--', c='r')
# 图例
plt.legend()
# 调整位置到左下角 且不显示边框
plt.legend(loc='lower left', frameon=False)
```

### 绘制参考区域

```python
x = np.linspace(0.05, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.axvspan(xmin=4, xmax=6, facecolor='r', alpha=0.3) # 垂直x轴
plt.axhspan(ymin=-0.2, ymax=0.2, facecolor='y', alpha=0.3);  # 垂直y轴
```

### 添加注释文字

```python
x = np.linspace(0.05, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.text(3.2, 0, 'sin(x)', weight='bold', color='r');
```

### 箭头标示

```python
x = np.linspace(0.05, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.annotate('maximum',xy=(np.pi/2, 1),xytext=(np.pi/2+1, 0.5),
             weight='bold',
             color='r',
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='r'));
```

```python
y.shape
```

```python
fig, ax = plt.subplots()

lines = []
styles = ['-', '--', '-.', ':']
x = np.linspace(0, 10, 1000)

for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2),
                     styles[i], color='black')
ax.axis('equal')

# 设置第一组标签
ax.legend(lines[:2], ['line A', 'line B'],
          loc='upper right', frameon=False)

# 创建第二组标签
from matplotlib.legend import Legend
leg = Legend(ax, lines[2:], ['line C', 'line D'],
             loc='lower right', frameon=False)
ax.add_artist(leg);
```

## 多子图

### 在指定位置创建指定大小子图

```python
# 在(0.65, 0.65)创建 (0.2， 0.2)大小子图
ax1 = plt.axes()
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2])

ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x), 'r')
ax2.grid()
```

```python
fig = plt.figure()
# 左下角的坐标 x:10%, y:80%
# 所占比例 宽度80%, 高度40%
ax1 = fig.add_axes([0.1, 0.8, 0.8, 0.4], ylim=(-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4], ylim=(-1.2, 1.2))

x = np.linspace(0, 10)
ax1.plot(np.sin(x));
ax1.set_title('ax1')
ax2.plot(np.cos(x));
ax2.set_title('ax2')
```

### 组合绘制大小不同的子图

```python
# 内容很简单 可以自己看看
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
plt.subplot(grid[0, 0])
plt.text(0.5, 0.5, '1')
plt.subplot(grid[0, 1:])
plt.text(0.5, 0.5, '2')
plt.subplot(grid[1, :2])
plt.text(0.5, 0.5, '3')
plt.subplot(grid[1, 2]);
plt.text(0.5, 0.5, '4')
```

### 显示一组二维数据的频度分布

```python
# 生成数据
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# 建立图像
fig = plt.figure(figsize=(6, 6))
# 其实是4*4个图
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
# 主图在垂直方向留出最后一个位置(-1)
# 主图在水平方向留出第一个位置(0)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# 画主图像的散点图
main_ax.scatter(x, y,s=3,alpha=0.2)
main_ax.set_xlabel('x')
main_ax.set_ylabel('y')

# 画图
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal')
y_hist.invert_xaxis()
```

## 对数据集的可视化

### 导入pandas

```python
import pandas as pd
```

### 导入数据

```python
df = pd.read_csv('./Pokemon.csv', index_col=0)
```

```python
df.head()
```

### 展示前五个数据的堆积柱状图

```python
# 提取相关数据
pokemon = df['Name'][:5]
hp = df['HP'][:5]
attack = df['Attack'][:5]
defense = df['Defense'][:5]
ind = [x for x, _ in enumerate(pokemon)]

# 从下往上顺序是 hp, attack, defense
# 通过bottom调整
plt.figure(figsize=(10,10))
plt.bar(ind, defense, width=0.8, label='Defense',
        color='blue', bottom=attack+hp)
plt.bar(ind, attack, width=0.8, label='Attack', 
        color='gold', bottom=hp)
plt.bar(ind, hp, width=0.8, label='Hp', 
        color='red')

# 设置坐标轴相关名称
plt.xticks(ind, pokemon)
plt.ylabel("Value")
plt.xlabel("Pokemon")
plt.legend(loc="upper right")
plt.title("5 Pokemon Defense & Attack & Hp")
```

### 展示前N=5个宝可梦的 Attack, HP 簇状柱状图

```python
N = 5
# 取出数据
pokemon_hp = df['HP'][:N]
pokemon_attack = df['Attack'][:N]
ind = np.arange(N) 
# 绘图
width = 0.35       
plt.bar(ind, pokemon_hp, width, label='HP')
plt.bar(ind + width, pokemon_attack, width,label='Attack')
# 添加坐标名和标题
plt.ylabel('Values')
plt.title('Pokemon Hp & Attack')
# 添加横轴刻度名称
plt.xticks(ind + width / 2, (df['Name'][:5]), rotation=45)
plt.legend(loc='best')
plt.show()
```

### 展示前N=5个宝可梦的堆积折线图

```python
N = 5
# 取出数据
pokemon_name = df['Name'][:N]
pokemon_hp = df['HP'][:N]
pokemon_attack = df['Attack'][:N]
pokemon_defense = df['Defense'][:N]
ind = np.arange(N)

labels = ["HP ", "Attack", "Defense"]

fig, ax = plt.subplots()
ax.stackplot(pokemon_name, pokemon_hp, pokemon_attack, pokemon_defense)
ax.legend(loc='upper left', labels=labels)
plt.xticks(rotation=90)
plt.show()
```

```python
N = 5
# 取出数据
pokemon_name = df['Name'][:N]
pokemon_hp = df['HP'][:N]
pokemon_attack = df['Attack'][:N]
pokemon_defense = df['Defense'][:N]

# 创建子项目共用横轴
fig, (ax1, ax2, ax3) = plt.subplots(3, sharex=True)

ax1.plot(pokemon_name, pokemon_hp, 'ko-')
ax1.set(ylabel='HP')

ax2.plot(pokemon_name, pokemon_attack, 'r.-')
ax2.set(ylabel='Attack')

ax3.plot(pokemon_name, pokemon_defense, ':')
ax3.set(ylabel='Defense')

# 总标题
fig.suptitle('3 subplots')

plt.show()
```

### 前15个的属性折线图

```python
plt.plot(df['HP'][:15], '-r',label='HP')
plt.plot(df['Attack'][:15], ':g',label='Attack')
plt.legend();
```

### 通过散点能展示三个值展示数据

```python
# 获取数据
x = df['Attack']
y = df['Defense']
colors = df['HP']

plt.scatter(x, y, c=colors, alpha=0.5)
plt.title('Scatter plot')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.colorbar();
```

### 攻击力分布直方图(bin=10)

```python
x = df['Attack']
num_bins = 10
plt.hist(x, num_bins, facecolor='blue', alpha=0.5)
plt.title('Histogram')
plt.xlabel('Attack')
plt.ylabel('Value')
plt.show()
```

### 饼图

```python
plt.figure(1, figsize=(8,8))
df['Type 1'].value_counts().plot.pie(autopct="%1.1f%%")
# 调整图例的位置
plt.legend(bbox_to_anchor=(1.4, 1))
plt.show()
```

### 属性统计柱状图

```python
ax = df['Type 1'].value_counts().plot.bar(figsize = (12,6), fontsize = 14)
ax.set_title("Pokemon Type 1 Count", fontsize = 20)
ax.set_xlabel("Pokemon Type 1", fontsize = 20)
ax.set_ylabel("Value", fontsize = 20)

plt.show()
```

### 展示Total项最高的10只宝可梦的相关系数矩阵

```python
import seaborn as sns

# 选出前10的宝可梦
top_10_pokemon=df.sort_values(by='Total',ascending=False).head(10)
# 计算各个属性之间的相关系数
corr=top_10_pokemon.corr()

# 构造图像(设置图像大小)
fig, ax=plt.subplots(figsize=(10, 6))
# 画热力图
sns.heatmap(corr, annot=True, cmap='rainbow_r')

plt.show()
```
