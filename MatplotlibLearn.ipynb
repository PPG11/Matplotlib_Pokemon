## 画线

### 基本绘图

```python
x = np.linspace(0, 10, 50)
plt.plot(x, np.sin(x))
# 网格
plt.grid()
# 坐标轴范围
plt.axis([-0.5, 10.5, -1.5, 1.5])
```

### 更改画线方式（圆圈点）

```python
plt.plot(x, np.sin(x), '-o');
```

### 单纯画散点

```python
plt.scatter(x, np.sin(x));
```

### 散点+面积表示

```python
# 种子
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
```

```python
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
# 展示右侧色阶
plt.colorbar();
```

### 误差条图

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
```

```python
plt.errorbar(x, y, yerr=dy, fmt='.k')
plt.plot(x, np.sin(x))
```

### 多个直方图及设置透明度

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(alpha=0.3, bins=40, density = True)

plt.hist(x1, **kwargs);
plt.hist(x2, **kwargs);
plt.hist(x3, **kwargs);
```

## 自定义图表元素

### plot线型

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x),'--');
```

### 坐标轴范围

```python
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x))
plt.ylim(-1.5, 1.5);
```

### 常用

```python
x = np.linspace(0.05, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='sin(x)')
# xy轴标签
plt.xlabel('variable x');
plt.ylabel('value y');
# 标题
plt.title('sin x');
# 水平参考线
plt.axhline(y=0.8, ls='--', c='r')
# 图例
plt.legend()
# 调整位置到左下角 且不显示边框
plt.legend(loc='lower left', frameon=False)
```

### 绘制参考区域

```python
x = np.linspace(0.05, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.axvspan(xmin=4, xmax=6, facecolor='r', alpha=0.3) # 垂直x轴
plt.axhspan(ymin=-0.2, ymax=0.2, facecolor='y', alpha=0.3);  # 垂直y轴
```

### 添加注释文字

```python
x = np.linspace(0.05, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.text(3.2, 0, 'sin(x)', weight='bold', color='r');
```

### 箭头标示

```python
x = np.linspace(0.05, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.annotate('maximum',xy=(np.pi/2, 1),xytext=(np.pi/2+1, 0.5),
             weight='bold',
             color='r',
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='r'));
```

```python
fig, ax = plt.subplots()

lines = []
styles = ['-', '--', '-.', ':']
x = np.linspace(0, 10, 1000)

for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2),
                     styles[i], color='black')
ax.axis('equal')

# 设置第一组标签
ax.legend(lines[:2], ['line A', 'line B'],
          loc='upper right', frameon=False)

# 创建第二组标签
from matplotlib.legend import Legend
leg = Legend(ax, lines[2:], ['line C', 'line D'],
             loc='lower right', frameon=False)
ax.add_artist(leg);
```

## 定义色阶

### 展示色阶

```python
x = np.linspace(0, 10, 1000)
I = np.sin(x) * np.cos(x[:, np.newaxis])

plt.imshow(I)
plt.colorbar();
```
